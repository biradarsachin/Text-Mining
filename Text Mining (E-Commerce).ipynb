```python
url = "https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1"
urls = []
```

```python
for var in range(1,18):
    url = ["https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=%i"%(var)]
    urls.append(url[0])
```

```python
def connect(ur):
    uClient = urlopen(ur)
    page_html = uClient.read()
    
    uClient.close()
    
    page_soup = soup(page_html,"html.parser")
    
    container = page_soup.find_all("div",{"class":"_2kHMtA"})
    return container
```

```python
con = connect(urls[0])
print(soup.prettify(con[1]))
```

```python
con_urls = connect(urls[0])
```

```python
for i in range(0,17):
    con = connect(urls[i])
    for container in con:
        product_name = container.img["alt"]
        print(product_name)
```

```python
for container in con:
    s_price = container.findAll("div",{"class":"_30jeq3 _1_WHN1"})
    actual_price = container.findAll("div",{"class":"_3I9_wc _27UcVY"})
    rating = container.findAll("div",{"class":"_3LWZlK"})
    p_rating = container.findAll("span",{"class":"_2_R_DZ"})
    offer = container.findAll("div",{"class":"_3Ay6Sb"})
    print(offer)
```

```python
def dataframe(containers):
    data = {"Product_name":[],"Sale price":[],"Actual Price":[],"Ratting":[],"Person Ratting":[],"Offer":[]}
    for i in range(0,15):
        con = connect(containers[i])
        for container in con:
            #Product Name:-
            product_name = container.img["alt"]
            data["Product_name"].append(product_name)
            
            #Sale price:-
            s_price = container.findAll("div",{"class":"_30jeq3 _1_WHN1"})
            s_p = s_price[0].text
            data["Sale price"].append(s_p)
            
            #Actual Price:-
            try:
                actual_price = container.findAll("div",{"class":"_3I9_wc _27UcVY"})
                a_p = actual_price[0].text
                data["Actual Price"].append(a_p)
            except IndexError:
                print("Same as sale price")
                data["Actual Price"].append("Same Price")
            
            #Rating:-
            try:
                ratting = container.findAll("div",{"class":"_3LWZlK"})
                r = ratting[0].text
                data["Ratting"].append(r)
            except IndexError:
                data["Ratting"].append("No Ratting")
            
            #No of person given ratting:-
            try:
                p_retting = container.findAll("span",{"class":"_2_R_DZ"})
                p_r = p_retting[0].text
                data["Person Ratting"].append(p_r)
            except IndexError:
                data["Person Ratting"].append("No Ratting")
                        
            
            #Offer:-
            try:
                offer = container.findAll("div",{"class":"_3Ay6Sb"})
                off = offer[0].text
                data["Offer"].append(off)
            except IndexError:
                data["Offer"].append("No offer this product")
            
        print("done",i)
        
    return data  
```

```python
dataset  =  dataframe(urls)
```

```python
df = pd.DataFrame(dataset)
```
